##데이터 불러오기

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
movies_data = pd.read_csv('/content/drive/MyDrive/Data_csv/movies.csv')
rating_data = pd.read_csv('/content/drive/MyDrive/Data_csv/ratings.csv')
```

##데이터 전처리

```python
#drop과 merge
rating_data.drop('timestamp', axis=1, inplace=True)
rating_data.head(2)
```

```python
user_movie_rating = pd.merge(rating_data, movies_data, on = 'movieId')
```

```python
user_movie_rating.head(2)
```

###pivot table 만들기
- data: 영화 평점 rating
- index: 영화 title
- columns: userId

(아이템 기반이므로, index = movies title, colomn = userId)

```python
#타이틀-유저아이디 pivot table
movie_user_rating = user_movie_rating.pivot_table('rating', index = 'title', columns='userId')
#유저아이디-타이틀 pivot table
user_movie_rating = user_movie_rating.pivot_table('rating', index = 'userId', columns='title')
```

```python
movie_user_rating.head()
```

```python
user_movie_rating.head()
```

###null값 채우기 

```python
#NaN 값 데이터에 fillna(0, inplace = True)로 null값 채우기 
movie_user_rating.fillna(0, inplace = True)
movie_user_rating.head()
```

## 유사도(similarity) 측정하기
- 코사인 유사도 사용. 
- scikitlearn 의 cosine similarity 지원
- argsort 로 유사도가 가장 높은 인덱스를 오름순차로 정렬
- 유사한 아이템끼리 추천하며, 여기서는 평점이 비슷한 아이템을 추천함

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 현재 평점이 data 이므로 이상태에서 코사인 유사도 사용
item_based_collabor = cosine_similarity(movie_user_rating)
item_based_collabor
```

```python
#Dataframe화 하기
item_based_collabor = pd.DataFrame(data = item_based_collabor, index = movie_user_rating.index, columns = movie_user_rating.index)
```

```python
item_based_collabor.head()
```

```python
#사용자가 시청한 영화가 마음에 들었을때, 그 영화과 비슷한 영화 추천 기능
def get_item_based_collabor(title):
  return item_based_collabor[title].sort_values(ascending=False)[1:7] 
  # item_based_collabor[title] 코사인 유사도 행렬에서 title 뽑아오기
  #sort_values(ascending=False)[1:7] 내림차순 후 본인 제외 앞 5개 가져오기
```

##결과

```python
get_item_based_collabor('"Great Performances" Cats (1998)')
```

```python
get_item_based_collabor('Fast and the Furious, The (2001)')
```
